```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import category_encoders as ce
```

```python
train= pd.read_csv('train_values.csv')
test= pd.read_csv('test_values.csv')
train_label =pd.read_csv('train_labels.csv')
```

```python
join_train = pd.merge(train, train_label, on='row_id')
join_train.head()
```

```python
join_train.isnull().sum()
```

```python
join_train.fillna(-999,inplace=True)
test.fillna(-999,inplace=True)
```

```python
x = join_train.drop('accepted',axis=1)
y = join_train.accepted
```

```python
x.dtypes
```

```python
#instantiate an encoder - here we use target()
#we also use a high vlue for smoothing in order to reduce overfitting

ce_target = ce.TargetEncoder(cols = ['loan_type','property_type','loan_purpose','occupancy','preapproval','msa_md','state_code','county_code','applicant_ethnicity','applicant_race','applicant_sex','lender','co_applicant'],smoothing = 100)
```

```python
# fit and transform to encoded data
ce_target.fit_transform(x,y)
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.15)
```

```python
#Scaling our features to reducing features with high values from having more influence than other features in our result
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
x_train = sc_X.fit_transform(x_train)
x_test = sc_X.transform(x_test)
```

```python
import xgboost as xgb
from xgboost.sklearn import XGBClassifier


xgb_model = xgb.XGBClassifier(objective="binary:logistic", random_state=42, learning_rate= 0.1,n_estimators= 200, max_depth = 5, gamma= 5, reg_alpha = 0.3)
xgb_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = xgb_model.predict(x_test)
ac = accuracy_score(y_test,y_pred)
print('Accuracy of XGboost model on test set: ',round(ac,2))
```

#### Tuning the learning rate and n_estimators

```python
import xgboost as xgb 
from xgboost.sklearn import XGBClassifier


xgb_model1 = xgb.XGBClassifier(objective="binary:logistic", random_state=42, learning_rate= 0.05,n_estimators= 500, max_depth = 5)
xgb_model1.fit(x_train, y_train)
```

```python
y_pred1 = xgb_model1.predict(x_test)
ac1 = accuracy_score(y_test,y_pred1)
print('Accuracy of XGBoost model1 set: ',round(ac1,2))
```
